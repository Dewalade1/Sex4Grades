```python
# Dependencies import
from nltk.corpus import stopwords
import pandas as pd
```

```python
# Columns name definition
cols = [
    'tweet_created_at',
    'tweet_text',
    'tweet_lang',
    'tweet_user_location',
    'tweet_user_verified',
    'tweet_source',
    'tweet_retweet_count',
    'tweet_favorite_count']
```

```python
# Parse the date and use just days.
# If the date field has time it will be droped.
def date_parser(serie):
    return serie.astype('datetime64[D]')

dataset_path = '../Data/Sex4Grades.csv'

# Dataset import.
# If the dataset already has the right columns name
# uncomment the second paramater into the import.
ds = pd.read_csv(dataset_path,
                 #usecols = cols,
                 date_parser=date_parser,
                 parse_dates=[0])
ds.columns = cols
ds.head()
```

```python
# Check right date parsing.
ds.info()
```

```python
stopwords_en = stopwords.words('english')
    
def clean_tweets(df, column):
    '''Using pandas built-in functions is more performance than python one.
    Remove RT old fashioned characters, urls, special characters and
    apply lower case to tweets.
    '''
    df.loc[:, column] = df[column].str.replace('^RT[\s]+', '')\
                                  .str.replace('https?:\/\/.*[\r\n]*', '')\
                                  .str.replace('[^\w\s]','')\
                                  .str.lower()
            
    return df

def remove_stopwords_from_tweet(tweet):
    return ' '.join(word for word in tweet.split() if word not in stopwords_en)

def filter_stopwords(df, column):
    df.loc[:, column] = df[column].apply(remove_stopwords_from_tweet)
    return df

def clean_tweet_source(df, column):
    df.loc[:, column] = df[column].str.split().str.get(-1)
    return df
```

```python
# Apply all the cleaning function using pandas pipe.
ds.pipe(clean_tweets, 'tweet_text')\
  .pipe(filter_stopwords, 'tweet_text')\
  .pipe(clean_tweet_source, 'tweet_source')
```

```python
ds.to_csv("sex4Grades_clean.csv")
```
